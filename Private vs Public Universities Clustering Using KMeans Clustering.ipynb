## Model Training

```python
# Import KMeans from SciKit Learn.
from sklearn.cluster import KMeans

# Create an instance of a KMeans() model named kmeans
kmeans = KMeans(n_clusters=2)

# Fit the model to all the data except for the Private label.
kmeans.fit(df.drop('Private', axis=1))
```

```python
# Create a new column for df called 'Cluster', which is a 1 for a Private school, and a 0 for a public school.
def converter(private):
    if private == 'Yes':
        return 0
    else:
        return 1

# Apply converter function to the private column
df['Cluster'] = df['Private'].apply(converter)

# Check the head of the df
df.head()
```

## Model Evaluation

```python
# Create a classification report for the model to see how ell the KMeans clustering worked without being given any labels.
from sklearn.metrics import classification_report
print(classification_report(df['Cluster'], kmeans.labels_))
```

```python
# Create a confusion matrix for the model to see how ell the KMeans clustering worked without being given any labels.
from sklearn.metrics import confusion_matrix
print(confusion_matrix(df['Cluster'], kmeans.labels_))
```

```python
# Create scatterplot of the Apps and Accept column in order to see difference between KMeans clustering results and Real labels 
fig, (ax1,ax2) = plt.subplots(1,2, sharey= True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(df.drop('Private', axis=1)['Apps'], 
            df.drop('Private', axis=1)['Accept'], 
            c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Real Labels')
ax2.scatter(df.drop('Private', axis=1)['Apps'][df['Cluster'] == 1], 
            df.drop('Private', axis=1)['Accept'][df['Cluster'] == 1], label = 'Public', c = 'red')
ax2.scatter(df.drop('Private', axis=1)['Apps'][df['Cluster'] == 0], 
            df.drop('Private', axis=1)['Accept'][df['Cluster'] == 0], label = 'Private', c = '#0450bf')
ax2.legend(bbox_to_anchor=(1, 0.8, 0.5, 0.2), loc=2)
```
